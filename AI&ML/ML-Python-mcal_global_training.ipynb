```python
import matplotlib.pyplot as plt
x = [1, 2, 3, 4, 5, 6]
y = [1, 4, 9, 16, 25, 36]
y1 = [1, 3, 6, 8, 10, 12]
plt.plot(x, y, label = 'First Line', color = 'blue')
plt.plot(x, y1, label = 'Second Line', color = 'red')
plt.legend()
plt.xlabel('xaxis')
plt.ylabel('yaxis')
plt.title('My First Chart')
plt.show()
```
